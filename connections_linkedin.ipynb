```python
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv("Connections.csv", header=2)
```

```python
file.head()
```

```python
if file.iloc[0, 0] == "First Name":
    file = file[1:]
```

```python
file.columns = ["First Name", "Last Name", "URL", "Email Address", "Company", "Position", "Connected On"]
```

```python
file.reset_index(drop=True, inplace=True)
file = file.drop(columns=["Email Address", "Position", "Connected On"])
```

```python
file.head()
```

```python
file.to_csv("Connections@linkedin.csv")
```

```python
file.dropna(inplace=True)
file.drop_duplicates(inplace=True)
```

```python
file = file[2:]
file
```

```python
file['name'] = file['First Name'] + ' ' + file['Last Name']
file.drop(columns=['First Name', 'Last Name'], inplace=True)
display(file.head())
```

```python
file.drop(columns=['Unnamed: 0.1','Unnamed: 0'], inplace=True)
display(file.head())
```

```python
file['Company'] = file['Company'].replace('Christ University, Bangalore', 'Christ University')
```

```python
file['Company'] = file['Company'].replace('NASA - National Aeronautics and Space Administration', 'NASA')
```

```python
file['Company'] = file['Company'].replace('CAPS- Centre For Academic And Professional Support', 'CAPS - Christ')
```

```python
file['Company'] = file['Company'].replace(
    'Centre for Placement and Career Guidance (CPCG) - CHRIST (Deemed to be University)',
    'CPCG - Christ'
)
```

```python
file['Company'] = file['Company'].replace(
    'CPCG - Christ', 'CPCG'
)
```

```python
file['Company'] = file['Company'].replace(
    'Samagra (CS Association)', 'Samagra - CS Club')
```

```python
top_companies = file['Company'].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
plt.bar(top_companies.index, top_companies.values)
plt.xlabel('Company')
plt.ylabel('Number of Connections')
plt.title('Top 10 Companies by Number of Connections')
plt.tight_layout()
plt.show()
```

```python
file['Position'] = file['Position'].apply(
    lambda x: 'Student' if isinstance(x, str) and any(keyword in x.lower() for keyword in ['intern', 'student', 'member', 'representative','designer', 'mentee', 'peer educator','volunteer']) else x
)
```

```python
file['Position'] = file['Position'].replace('Founder', 'CEO')
file['Position'] = file['Position'].replace('Chief Technology Officer', 'CTO')
file['Position'] = file['Position'].apply(lambda x: 'Software Engineer' if isinstance(x, str) and ('software engineer' in x.lower() or 'developer' in x.lower() or 'sde' in x.lower() or 'technical' in x.lower()) else x)
```

```python
file['Position'] = file['Position'].apply(lambda x: 'Analyst' if isinstance(x, str) and ('analyst' in x.lower() or 'data' in x.lower()) else x)
file['Position'] = file['Position'].apply(lambda x: 'Professor' if isinstance(x, str) and 'professor' in x.lower() else x)
file['Position'] = file['Position'].apply(lambda x: 'Executive' if isinstance(x, str) and ('executive' in x.lower() or 'associate' in x.lower()) else x)
file['Position'] = file['Position'].apply(lambda x: 'Scientist' if isinstance(x, str) and 'scientist' in x.lower() else x)
```

```python
import seaborn as sns

top_positions = file['Position'].value_counts().nlargest(10).reset_index()
top_positions.columns = ['Position', 'Count']

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=top_positions, x='Position', y='Count', palette='Set1')

plt.title('Top 10 Positions by Number of Connections', fontsize=14)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Number of Connections', fontsize=12)

plt.tight_layout()
plt.show()
```
